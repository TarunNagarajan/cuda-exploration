# Intro to GPUs - 19/01/2025

A core is the basic computational unit of a processor (CPU or GPU) that can execute instructions independently.


image_2025-01-19_094036852.png

## CPU Example:
You open a web browser, and one core processes the browser's code while another core processes your media player running in the background.
A CPU is great for serial tasks (tasks that must be executed step-by-step, like program logic).

## GPU Example:
When rendering graphics in a video game, a GPU uses thousands of cores to calculate pixel colors, lighting, and physics for every frame simultaneously.
A GPU is ideal for parallel tasks (tasks that can run independently, like matrix multiplications in machine learning).



Each CUDA core (GPU core) executes a single thread.
Multiple cores form a Streaming Multiprocessor (SM), which manages groups of threads.
Threads are divided into blocks, and blocks are organized into a grid.

A clock cycle is the smallest unit of time in a processor.
It determines when the core performs operations like fetching instructions, executing them, or moving data.
Higher clock speed means a core can perform more operations in a given amount of time.
